```python
# imports libraries
import requests
import pandas as pd
import time

# retrieves the last 30 days of weather data for Louisville 
url = "https://community-open-weather-map.p.rapidapi.com/climate/month"

querystring = {"q":"Louisville"}

headers = {
	"X-RapidAPI-Host": "community-open-weather-map.p.rapidapi.com",
	"X-RapidAPI-Key": "7a595a9e5emsh435529e7efd4996p178108jsnc732a28cd5fe"
}

response = requests.request("GET", url, headers=headers, params=querystring)

data = response.json()['list']

# assigns weather data to pandas dataframe
df = pd.DataFrame(data=data)

# prints raw weather data for the last 30 days in pandas dataframe
print('RAW WEATHER DATA FOR THE LAST 30 DAYS IN LOUISVILLE, KY')
print('')
print(df.to_string())
print('')
print('')

# creates seprate split_temp pandas dataframe, adds "dt" column and splits initial temperature data into five different columns
split_temp = pd.DataFrame(df['temp'].tolist(), columns = ['dt','average','average_max','average_min','record_max','record_min'])

# prints split temperature dataframe 
print('Original temperature data was split into five columns')
print('')
print(split_temp.to_string())
print('')
print('')

# deletes initial "temp" column from the original "df" dataframe
del df['temp']

# adds "dt" identifier column from "df" dataframe to "split_temp" dataframe (*dt identifier will be used to merage two dataframes together)
split_temp.dt = df.dt

# merges "df" and "split_temp" into "df_merge" dataframe
df_merge = pd.merge(df,split_temp)

# prints replaced split temperature dataframe that is now merged into the original dataframe
print('MODIFIED WEATHER DATA FOR THE LAST 30 DAYS IN LOUISVILLE, KY')
print('Please note that the original temperature column was deleted and that the split temperature dataframe was added to the original raw weather data')
print('')
print(df_merge.to_string())
print('')
print('')

# converts temperature values from Kelvin to Fahrenheit using formula °F = (K − 273.15) × 9/5 + 32
df_merge.average = ((df_merge.average - 273.15)*(9/5)+32)
df_merge.average_max = ((df_merge.average_max - 273.15)*(9/5)+32)
df_merge.average_min = ((df_merge.average_min - 273.15)*(9/5)+32)
df_merge.record_max = ((df_merge.record_max - 273.15)*(9/5)+32)
df_merge.record_min = ((df_merge.record_min - 273.15)*(9/5)+32)

# changes column names for average, average_max, average_min, record_max, record_min and adds _temp to indicate temperature values
df_merge.rename(columns = {'average':'average_temp','average_max':'average_max_temp','average_min':'average_min_temp','record_max':'record_max_temp','record_min':'record_min_temp'}, inplace=True)

# prints modified weather data with temperatures in Fahrenheit instead of Kelvin
print('MODIFIED WEATHER DATA FOR THE LAST 30 DAYS IN LOUISVILLE, KY - KELVIN TO FAHRENHEIT CONVERSION')
print('Please note that the original temperature values were in Kelvin and that in this dataframe they have been converted to Fahrenheit')
print('')
print(df_merge.to_string())
print('')
print('')

# prints modified weather data only for days that had winds higher than 3 miles per hour
print('DAYS FOR THE LAST 30 DAYS IN LOUISVILLE THAT HAD WINDS HIGHER THAN 3 MPH')
print('')
print(df_merge.query('wind_speed > 3').to_string())
print('')
print('')

#prints average humidity for the last month
avg_humidity = int(df_merge.humidity.mean())
avg_humidity = str(avg_humidity)
print ('Average humidity for the last 30 days in Louisville was '+ avg_humidity + ' percent.')
print('')
print('')

#prints average wind speed for the last month
avg_wind_speed = int(df_merge.wind_speed.mean())
avg_wind_speed = str(avg_wind_speed)
print('Average wind speed for the last 30 days in Louisville was ' + avg_wind_speed + ' miles per hour.')
print('')
print('')

#print average temperature for the last month
mean_avg_temp = int(df_merge.average_temp.mean())
mean_avg_temp = str(mean_avg_temp)
print('Mean daily temperature for the last 30 days in Louisville was ' + mean_avg_temp + ' degrees Fahrenheit.')


```
